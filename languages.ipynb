```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/List_of_Wikipedias"
    
# Add headers to avoid being blocked
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

response = requests.get(url, headers=headers)
response.raise_for_status()

soup = BeautifulSoup(response.content, 'html.parser')

# Find the main table with Wikipedia statistics
# The table has columns: Language, Native name, Code, Articles, etc.
table = soup.find('table')
```

```python
counts = {}
rows = table.find_all('tr')
for row in rows:
    try:
        cells = row.find_all('td')
        code = cells[3].text.upper()
        count = int(cells[4].text.replace(',', ''))
        counts[code] = count
    except:
        pass

counts
```

```python
import json

with open('data/language_counts.json', 'w+') as f:
    json.dump(counts, f)
```
